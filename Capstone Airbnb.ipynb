## **Analisis Data**

Analisis ini dilakukan dengan melihat popularitas sebagai patokan/acuan/tolak ukurnya. Tingkat popularitas ini dilihat dari jumlah review pengguna terhadap listing/properti yang telah disewa. Ada beberapa hal yang akan dilakukan sebagai bagian dari proses analisa data:

1. **Berdasarkan Lokasi**:
    - Pengaruh neighbourhood terhadap jumlah review
    <br><br>
2. **Berdasarkan Host**:
    - Pengaruh host dan jumlah properti host terhadap jumlah review
    <br><br>
3. **Berdasarkan Tipe Listing**:
    - Pengaruh tipe properti/listing terhadap jumlah review

Analisis data
1. **Berdasarkan Lokasi**:
    - Pengaruh neighbourhood terhadap jumlah review
    <br><br>

```python
# Kelompokkan berdasarkan lokasi lalu jumlahkan review
top_locations = df.groupby("neighbourhood")["number_of_reviews"].sum().sort_values(ascending=False).head(10)

# Tiap lokasi memiliki perbedaan warna bar chart
colors = plt.cm.tab10(range(len(top_locations)))

# Plot bar chart
plt.figure(figsize=(12,6))
top_locations.plot(kind="bar", color=colors)

plt.title("Top 10 Lokasi Airbnb dengan Total Jumlah Review Terbanyak di Bangkok", fontsize=14)
plt.xlabel("Lokasi", fontsize=12)
plt.ylabel("Jumlah Review", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Ubah kolom tanggal review ke bentuk datetime
df["last_review"] = pd.to_datetime(df["last_review"], errors="coerce")

# Tentukan batas waktu: 12 bulan terakhir dari data
max_date = df["last_review"].max()
min_date = max_date - pd.DateOffset(months=12)

# Filter hanya data 12 bulan terakhir
df_last12 = df[(df["last_review"] >= min_date) & (df["last_review"] <= max_date)]

# Hitung total jumlah review per lokasi
top10_last12 = (
    df_last12.groupby("neighbourhood")["number_of_reviews"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

# Plot bar chart
plt.figure(figsize=(12,6))
bars = plt.bar(top10_last12["neighbourhood"], top10_last12["number_of_reviews"], color=plt.cm.tab20.colors)
plt.title("Top 10 Lokasi Airbnb dengan Total Jumlah Review Terbanyak 12 Bulan Terakhir di Bangkok")
plt.xlabel("Lokasi")
plt.ylabel("Jumlah Review")
plt.xticks(rotation=30, ha="right")
plt.show()
```

Analisis data

2. **Berdasarkan Host**:
    - Pengaruh host dan jumlah properti host terhadap jumlah review
    <br><br>

```python
# Hitung total review per host
top10_hosts = (
    df.groupby("host_name")["number_of_reviews"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

# Plot bar chart
plt.figure(figsize=(12,6))
bars = plt.bar(top10_hosts["host_name"], top10_hosts["number_of_reviews"], color=plt.cm.Set3.colors)
plt.title("Top 10 Host Airbnb dengan Total Jumlah Review Terbanyak di Bangkok")
plt.xlabel("Nama Host")
plt.ylabel("Jumlah Review")
plt.xticks(rotation=30, ha="right")
plt.show()
```

```python
# Hitung jumlah listing (list_count) dan total review per host
df_host_summary = (
    df.groupby("host_name")
    .agg(
        list_count=("id", "count"),              # hitung jumlah listing (pakai id listing)
        number_of_reviews=("number_of_reviews", "sum")  # jumlahkan semua review
    )
    .reset_index()
    .sort_values(by="number_of_reviews", ascending=False)
).head(10)

df_host_summary
```

```python
# Ubah kolom last_review dalam format datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Cari tanggal maksimum di dataset
max_date = df['last_review'].max()

# Filter hanya 12 bulan terakhir
last_12_months = max_date - pd.DateOffset(months=12)
df_last_year = df[df['last_review'] >= last_12_months]

# Hitung total review per host
host_reviews = (
    df_last_year.groupby("host_name")["number_of_reviews"]
    .sum()
    .reset_index()
)

# Ambil 10 host teratas
top10_hosts = host_reviews.sort_values(by="number_of_reviews", ascending=False).head(10)

# Plot
plt.figure(figsize=(12,6))
sns.barplot(
    data=top10_hosts,
    x="host_name",
    y="number_of_reviews",
    hue='host_name',
    legend=False,
    dodge=False,
    palette="Set2"
)

plt.title("Top 10 Host Airbnb dengan Total Jumlah Review Terbanyak di Bangkok (12 Bulan Terakhir)", fontsize=14)
plt.xlabel("Nama Host")
plt.ylabel("Jumlah Review")
plt.xticks(rotation=30)
plt.show()
```

```python
# Pastikan kolom last_review dalam format datetime
df['last_review']=pd.to_datetime(df['last_review'], errors='coerce')

# Cari tanggal maksimum di dataset
max_date=df['last_review'].max()

# Filter hanya 12 bulan terakhir
last_12_months=max_date-pd.DateOffset(months=12)
df_last_year=df[df['last_review'] >= last_12_months]

# Hitung total review per host 
host_reviews=(
    df_last_year.groupby('host_name')['number_of_reviews']
    .sum()
    .reset_index()
)

# Hitung jumlah listing per host (list_count)
host_list_count=(
    df_last_year.groupby('host_name')['name']
    .count()
    .reset_index()
    .rename(columns={'name':'list_count'})
)

# Gabungkan hasil review dan list_count
host_summary=pd.merge(host_reviews, host_list_count, on='host_name')

#Ambil 10 host teratas berdsasarkan jumlah review
top10_hosts=host_summary.sort_values(by='number_of_reviews', ascending=False).head(10)
top10_hosts
```

```python
#cek distribusi list_count
stats, pval=normaltest(df_host_summary['list_count'].head(20))
if pval<=0.05:
    print ('tidak normal') # Ha
else:
    print('distribusi normal') # H0
```

```python
#cek distribusi number_of_reviews
stats, pval=normaltest(df_host_summary['number_of_reviews'])
if pval<=0.05:
    print ('tidak normal') # Ha
else:
    print('distribusi normal') # H0
```

```python
#cek distribusi number_of_reviews_ltm
stats, pval=normaltest(df_host_summary['number_of_reviews'])
if pval<=0.05:
    print ('tidak normal') # Ha
else:
    print('distribusi normal') # H0
```

3. **Berdasarkan Tipe Listing**:
    - Pengaruh tipe listing terhadap jumlah review

```python
# Hitung total review per tipe listing
reviews_per_type = df.groupby('room_type')['number_of_reviews'].sum().reset_index()

# Plot bar chart
plt.figure(figsize=(8,6))
sns.barplot(
    x='room_type',
    y='number_of_reviews',
    data=reviews_per_type,
    hue='room_type',
    palette='viridis',
    dodge=False,  #biar tidak muncul double bar
    legend=False
)

plt.title("Jumlah Review Berdasarkan Tipe Listing")
plt.xlabel("Tipe Listing")
plt.ylabel("Jumlah Review")
plt.show()
```

```python
# Filter data 12 bulan terakhir
last_date = df['last_review'].max()
start_date = last_date - pd.DateOffset(months=12)
df_last12 = df[df['last_review'] >= start_date]

# Hitung jumlah review per tipe listing
reviews_last12 = df_last12.groupby('room_type')['number_of_reviews'].sum().reset_index()

# Plot bar chart
plt.figure(figsize=(8,6))
sns.barplot(
    x='room_type',
    y='number_of_reviews',
    data=reviews_last12, 
    hue='room_type',
    palette='Set2',
    dodge=False,
    legend=False
)

plt.title("Jumlah Review Berdasarkan Tipe Listing Dalam 12 Bulan Terakhir")
plt.xlabel("Tipe Listing")
plt.ylabel("Jumlah Review")
plt.show()
```

```python
# Hitung jumlah review per lokasi dan tipe listing
reviews_by_loc_type = (
    df.groupby(['neighbourhood', 'room_type'])['number_of_reviews']
      .sum()
      .reset_index()
)

# Ambil 10 lokasi dengan jumlah review terbanyak
top10_loc = (
    reviews_by_loc_type.groupby('neighbourhood')['number_of_reviews']
    .sum()
    .nlargest(10)
    .index
)

# Filter hanya top 10 lokasi
reviews_top10 = reviews_by_loc_type[reviews_by_loc_type['neighbourhood'].isin(top10_loc)]

# Plot grouped bar chart
plt.figure(figsize=(12,6))
sns.barplot(
    x='neighbourhood', 
    y='number_of_reviews', 
    hue='room_type', 
    data=reviews_top10
)

plt.title("Tipe Listing di Top 10 Lokasi")
plt.xlabel("Lokasi")
plt.ylabel("Jumlah Review")
plt.xticks(rotation=45)
plt.legend(title='Room Type')
plt.show()
```
